### Classification: decision boundaries and parameter selection

This exercise focuses on two main aspects: What kind of decision boundaries is produced by different classifiers? And how do these boundaries behave for different parameter choices?

In order to investigate this, we use the 'fifa_balanced.csv' dataset once again. Again we try to predict the 'Position', but this time only using only two columns, namely 'Interceptions' and 'Standing tackle',  as input.

```python
data_complete = pd.read_csv('data/fifa_balanced.csv')

target = 'Position'
input_columns = ['Interceptions', 'Standing tackle']
```

Split the data into X and y data and use a LabelEncoder to transform the classification target into a numerical value. Afterwards, use train_test_split method to create a training set (40%) and a test set (60%).

```python
# TODO: label-encode and split into train and test dataset

# input data
X = data_complete[input_columns]

# class labels
y = data_complete['Position']
le = LabelEncoder().fit(list(y))
y_encode = le.transform(list(y))

validation_size = 0.60
X_train, X_test, y_train, y_test = train_test_split(X, y_encode, test_size=validation_size)
```

At first, have a look at the Decision Tree classifier. Train models, using 1, 4 and 128 as different parameters for the maximum depth of the tree. For each parameter, print the test accuracy and plot the decision boundaries in a scatter plot for test and train data. For creating the plot, use the provided 'visualize_boundaries' method. As paramters it takes:
- the classifier
- the train input data as DataFrame
- the train target as Series/DataFrame
- the test input data as DataFrame
- the test target as Series/DataFrame

As output it returns a Figure object, which can be displayed using 'plt.show(figure)'.

```python
from sklearn.tree import DecisionTreeClassifier

depths = [1, 4, 128]

# TODO: train/test decision trees with different depths, visualize decision boundaries and print accuracy
for i in depths:
    tree = DecisionTreeClassifier(max_depth=i).fit(X_train, y_train)
    tree_predictions = tree.predict(X_test)
    print("DecisionTree accuracy with max_depth {0}".format(i),": " ,accuracy_score(y_test, tree_predictions))
    figure = visualize_boundaries(tree, X_train, y_train, X_test, y_test)
    plt.show(figure)
```

Now we do the same for a K-Nearst-Neighbors classifier. This time we change the number of neighbors, taken into considerstion for classification, using the values 1, 20, and 200 (200 could take a while). Again you can use the provided 'visualize_boundaries' mathod as stated aboth.

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = [1, 20, 200]

# TODO: train/test knn with different numbers of neighbors into account, visualize decision boundaries and print accuracy
for i in neighbors:
    knn = KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train)
    knn_predictions = knn.predict(X_test)
    print("KNeighborsClassifier accuracy with n_neighbors {0}".format(i),": ",accuracy_score(y_test, knn_predictions))
    figure = visualize_boundaries(knn, X_train, y_train, X_test, y_test)
    plt.show(figure)
```

Last of all, we use a Multilayer-Perceptron (neural network classifier) as third classifier and change the number of neurons in the hidden layer. Use (), (10,) and (200,) as parameters (no hidden layer at all, 10 neurons, 200 neurons). Once again print the test accuracy and and plot the decision boundaries.

**Hint:** Do multiple runs because the results can differ a lot for some paramter settings.

```python
from sklearn.neural_network import MLPClassifier

layers = [(), (10,), (200,)]

# TODO: train/test MLP with different structures, visualize dicision boundaries and print accuracy
for i in layers:
    hidden_layer_sizes=i
    ann = MLPClassifier(hidden_layer_sizes=i).fit(X_train, y_train)
    ann_predictions = ann.predict(X_test)
    print("MLPClassifier accuracy with hidden layer size {0}".format(i),": ",accuracy_score(y_test, ann_predictions))
    figure = visualize_boundaries(ann, X_train, y_train, X_test, y_test)
    plt.show(figure)
```

### Naive Bayes classifier

In order to repeat the concept of a Naive Bayes classifier, complete the given class 'MyNaiveBayes'. For a detailed explanation have a look at:
- http://scikit-learn.org/stable/modules/naive_bayes.html
- https://en.wikipedia.org/wiki/Naive_Bayes_classifier

**Hints:** 
- think about a suitable data structure for saving conditional and marginal probabilities (DataFrames, dictionaries consisting of DataFrame, ...)
- also use Pandas and Numpy data types for intermediate resuls and always look for functions that are already provided by Pandas and Numpy (argmax, unique, ...)
- if you do not come up with very efficient solutions, at least come up with some solution ;) 

```python
class MyNaiveBayes():
    def __init__(self):
        pass

    def fit(self, X, y):
        count_sample = X.shape[0]
        separated = [[x for x, t in zip(X, y) if t == c] for c in np.unique(y)]
        self.class_log_prior_ = [np.log(len(i) / count_sample) for i in separated]
        count = np.array([np.array(i).sum(axis=0) for i in separated]) 
        self.feature_log_prob_ = np.log(count / count.sum(axis=1)[np.newaxis].T)
        return self
     
    def predict_log_proba(self, X):
        return [(self.feature_log_prob_ * x).sum(axis=1) + self.class_log_prior_ for x in X]

    def predict(self, X):
        return np.argmax(self.predict_log_proba(X), axis=1)
    
    def score(self, X, y):
        prediction = self.predict(X)
        return (np.array(prediction) == np.array(y)).sum()/len(y)
```

For testing your implementation, test and train your classifier on the given 'fifa_discrete' dataset, which is a discretized version of the known 'fifa_balanced' dataset. Use 60% as training data. 

(the achieved accuracy should be > 95%)

```python
data = pd.read_csv('data/fifa_discrete.csv')
target = 'Position'
input_columns = [col for col in data.columns if col!=target]

y = data[target]
X = data.loc[:, input_columns]
y = y.as_matrix()
X = X.as_matrix()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)


# TODO: train and test MyNaiveBayes and print accuracy
nb = MyNaiveBayes().fit(X_train,y_train)
nb_score = nb.score(X_test,y_test)
print(nb_score)
```
